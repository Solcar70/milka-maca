### Marco Teórico

Actualmente existen diferentes metodolgías y acercamientos para poder analizar conjutos de datos, ya sea para poder realizar una clasificación o una regresión. Por dicha razón es que este proyecto se va a enfocar en dos métodos, Random Forest y MLP (Multilayer Perceptron).

Para entender el funcionamiento de Random Forest se va a iniciar con una explicación introductoria a los árboles de decisión, estos constituyen uno de los algoritmos más utilizados en el aprendizaje supervisado, tanto para clasificación como para regresión. Su estructura se asemeja a un árbol jerárquico: los nodos internos representan condiciones sobre atributos, las ramas reflejan los resultados de esas condiciones y las hojas corresponden a la salida final [8].

El objetivo del algoritmo es dividir recursivamente los datos en subconjuntos más homogéneos según una métrica de impureza o de error, donde las métricas más utilizadas en el campo son la entropía y el índice Gini.

Inciando con la entropía, esta mide el grado de desorden en un conjunto y se define como:

$$
H(S) = - \sum_{i=1}^{k} p_i \log_2(p_i)
$$

donde $p_i$ es la proporción de elementos de la clase $i$ en el conjunto $S$, y $k$ el número de clases. Cabe destacar que un nodo es puro si su entropía es 0.

La otra métrica común es el índice Gini:

$$
Gini(S) = 1 - \sum_{i=1}^{k} p_i^2
$$

que mide la probabilidad de clasificar erróneamente un elemento si se asigna aleatoriamente según la distribución de clases.

La calidad de una división se evalúa mediante la ganancia de información (Information Gain), que mide la reducción de impureza obtenida al dividir con un atributo $A$:

$$
IG(S, A) = H(S) - \sum_{v \in Valores(A)} \frac{|S_v|}{|S|} H(S_v)
$$

donde $S_v$ es el subconjunto de $S$ en el que el atributo $A$ toma el valor $v$.

En problemas de regresión, en lugar de impureza de clases se utiliza el error cuadrático medio (MSE):

$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

donde $y_i$ es el valor real y $\hat{y}_i$ la predicción promedio en el nodo.

Ahora que ya se tiene definido qué es un árbol de decisión y el funcionamiento del mismo, se puede pasar al método Random Forest que el método a usar en este proyecto. Este algoritmo surge precisamente para superar las limitaciones de un único árbol, el cual puede ser inestable y propenso al sobreajuste. Random Forest consiste en construir múltiples árboles de decisión de forma independiente y luego combinar sus resultados para obtener una predicción más robusta y precisa [9].  

Cada árbol se entrena sobre una muestra aleatoria del conjunto de datos original y en cada nodo solo se consideran algunas de las variables. Este funcionamiento aleatorio genera diversidad entre los árboles y reduce la correlación entre sus errores. El resultado final se puede obtener mediante votación en clasificación y mediante el promedio en regresión, esto depende del acercamiento que se esté utilizando.

- Clasificación:

$$
\hat{y} = \text{mode}\{h_1(x), h_2(x), ..., h_B(x)\}
$$

- Regresión:

$$
\hat{y} = \frac{1}{B} \sum_{b=1}^{B} h_b(x)
$$

donde $h_b(x)$ es la predicción del árbol $b$ y $B$ el número total de árboles en el bosque.

Este enfoque reduce la varianza de los árboles individuales sin aumentar de forma importante el sesgo, lo que hace a Random Forest un modelo robusto frente a datos complejos. Cabe destacar que este modelo sacrifica interpretabilidad, ya que al tener muchos árboles funcionando al mismo tiempo resulta difícil seguir el razonamiento de esos cientos de árboles y visualizar una única regla de decisión, sin embargo esto genera que se pueda tener mayor capacidad predictiva y estabilidad. En lugar de depender de un solo camino de decisiones el modelo combina múltiples perspectivas, lo que permite capturar patrones más complejos en los datos y minimizar los efectos del ruido o de variables irrelevantes.  

Además al promediar los resultados de varios árboles, Random Forest tiende a generalizar mejor y a reducir el riesgo de sobreajuste, un problema común en los árboles individuales. Si bien no permite explicar de forma sencilla cada predicción existen métricas como la importancia de las variables (feature importance) que ofrecen una visión global de cuáles factores influyen más en el modelo. Este equilibrio entre precisión y capacidad de interpretación parcial lo convierte en una herramienta especialmente adecuada para problemas aplicados en agricultura, como el análisis de la producción de café a nivel regional [10].  

Además de Random Forest, también se va a implemntar un Perceptrón Multicapa (MLP) el cuál es una de las estructuras más representativas de las redes neuronales artificiales. Los MLP son utilizados para aprendizaje profundo supervisado y es especialmente útil para las tareas de clasificación y regresión donde las relaciones no son necesariamente lineales, por eso para el caso del análisis de los datos de café en Costa Rica van a ser sumamente útiles. [11]

Se sabe que los MLP están compuestos por al menos 3 capas, una de entrada, una o más capas ocultas y una capa de salida. Se dice que el sistema está fuertemente conectado cuando cada una de las neuronas de una capa determinada está conectada a todas las neuronas de la capa siguiente. En este caso los datos fluyen hacia adelante a través de la red, y los errores se propagan hacia atrás durante el proceso de entrenamiento. Este último paso va a ajustando los pesos de la red en función del gradiente del error. [12]

La operación básica de una neurona consiste en aplicar una transformación lineal seguida de una función de activación no lineal:

$$
z = \sum_{i=1}^{n} w_i x_i + b
$$

$$
a = f(z)
$$

donde $x_i$ son las entradas, $w_i$ los pesos, $b$ es el sesgo (bias), $z$ es la combinación lineal, y $f(\cdot)$ es una función de activación, como ReLU, sigmoide o tangente hiperbólica.

El entrenamiento en los MLP funciona minimizando una función de pérdida, un procedimiento similar al que se realiza al errir cuadrático medio en regresión o entropía en árboles de decisión. En este caso se utilizan técnicas como el descenso de gradiente estocástico (SGD) aunque también existen técnicas más avanzadas como Adam. [13]

En el caso de los MLP, estos no son tan interpretables como los árboles de decisión, sin embargo tienen una alta capacidad de modelar relaciones complejas y no lineales entre variables lo que los hace apropiados para modelar el comportamiento del café en las diferentes zonas del país.

```python
# ====== bloque imports: librerías principales ======

# NumPy / Pandas
import numpy as np
import pandas as pd

# Visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Utilidades generales
import random
from copy import deepcopy
from IPython.display import display   # para mostrar tablas y objetos en Jupyter
import time


# PyTorch (para implementar el MLP)
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

# Scikit-learn (versión >= 1.4)
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error,      # se mantiene como apoyo para MSE
    root_mean_squared_error  # API moderna para RMSE
)
```

```python
# ====== bloque utilidades: funciones generales ======

def set_seed(seed: int = 42):
    # Fijar la semilla para reproducibilidad en random, numpy y torch
    import os
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False


def make_loaders(X_tr, y_tr, X_va, y_va, batch_train: int = 512, batch_eval: int = 1024):
    # Crear DataLoaders de PyTorch para entrenamiento y validación
    Xtr_t = torch.from_numpy(np.asarray(X_tr)).float()
    Xva_t = torch.from_numpy(np.asarray(X_va)).float()
    ytr_t = torch.from_numpy(np.asarray(y_tr)).float()
    yva_t = torch.from_numpy(np.asarray(y_va)).float()

    train_dl = DataLoader(TensorDataset(Xtr_t, ytr_t), batch_size=batch_train, shuffle=True)
    val_dl   = DataLoader(TensorDataset(Xva_t, yva_t), batch_size=batch_eval, shuffle=False)
    return train_dl, val_dl
```

```python
# ====== bloque preprocesamiento: función Data_Preprocess ======
# Limpia nulos, separa X/y, aplica OneHot a categóricas y StandardScaler a numéricas,
# realiza split train/test y devuelve matrices ya transformadas.

def Data_Preprocess(df: pd.DataFrame,
                    numeric_columns: list,
                    categoric_columns: list,
                    target_column: str,
                    test_size: float = 0.20,
                    random_state: int = 42):
    # Eliminar filas con nulos y reindexar
    df = df.dropna().reset_index(drop=True)
    print("\nShape después de limpieza:", df.shape)

    # Remover el target de las listas si aparece
    if target_column in categoric_columns: categoric_columns.remove(target_column)
    if target_column in numeric_columns:   numeric_columns.remove(target_column)

    # Separar X (features) e y (target continuo)
    X = df[categoric_columns + numeric_columns].copy()
    y_raw = df[target_column]

    # Asegurar target numérico (regresión)
    if np.issubdtype(y_raw.dtype, np.number):
        y = y_raw.astype(float).values
        print("Target numérico usado directamente (regresión).")
    else:
        y = pd.to_numeric(y_raw, errors="coerce").values
        if np.isnan(y).any():
            raise ValueError("Target no convertible a float para regresión.")
        print("Target convertido a float desde texto (regresión).")

    # Preprocesador: OneHot para categóricas + StandardScaler para numéricas
    column_encoder = ColumnTransformer(
        transformers=[
            ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop=None), categoric_columns),
            ("num", StandardScaler(), numeric_columns),
        ],
        remainder="drop"
    )

    # Split train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    # Ajustar en TRAIN y transformar TRAIN/TEST
    X_train_enc = column_encoder.fit_transform(X_train)
    X_test_enc  = column_encoder.transform(X_test)

    return X_train_enc, X_test_enc, y_train, y_test
```

### Modelo Random Forest

```python
# ====== bloque random forest: clase RFModule ======
# Implementa RandomForestRegressor con búsqueda de n_estimators usando validación cruzada (R²).
# Incluye ajuste final y reporte con métricas de regresión (R², MAE, RMSE).

class RFModule:
    def __init__(self, seed: int = 42):
        # Inicializar parámetros
        self.seed = seed
        self.model = None
        self.best_n = None

    def cv_n_estimators(self, X, y, n_list=(50,100,200,300,500), k: int = 5):
        # Búsqueda de la mejor cantidad de árboles mediante K-Fold CV
        cv = KFold(n_splits=k, shuffle=True, random_state=self.seed)
        results = []
        for n in n_list:
            rf = RandomForestRegressor(
                n_estimators=n,
                random_state=self.seed,
                n_jobs=-1
            )
            scores = cross_val_score(rf, X, y, cv=cv, scoring="r2", n_jobs=-1)
            results.append((n, float(np.mean(scores)), float(np.std(scores))))
        # Seleccionar el n_estimators con mayor R² promedio
        self.best_n = max(results, key=lambda t: t[1])[0]
        return results, self.best_n

    def fit(self, X, y, n_estimators=None):
        # Entrenamiento en todo el set de entrenamiento con n_estimators óptimo
        n = n_estimators if n_estimators is not None else (self.best_n or 100)
        self.model = RandomForestRegressor(
            n_estimators=n,
            random_state=self.seed,
            n_jobs=-1
        )
        self.model.fit(X, y)
        return self

    def report(self, X_test, y_test, title="RandomForestRegressor (TEST)"):
        # Evaluación con métricas de regresión
        y_pred = self.model.predict(X_test)
        r2  = r2_score(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        rmse = float(root_mean_squared_error(y_test, y_pred))  # sklearn >= 1.4

        print(f"\n=== {title} ===")
        print(f"R²   = {r2:.3f}")
        print(f"MAE  = {mae:.3f}")
        print(f"RMSE = {rmse:.3f}")

        return y_pred, {"R2": r2, "MAE": mae, "RMSE": rmse}
```

### Modelo MLP 3 capas

```python
# ====== bloque mlp regresión: clases MLPRegressor y MLPRegressorModule ======
# Implementa un MLP para REGRESIÓN en PyTorch con:
# - Arquitectura configurable (capas ocultas, dropout)
# - Early stopping por RMSE en validación
# - Validación cruzada K-Fold
# - Scheduler opcional ReduceLROnPlateau
# - Reporte con R², MAE y RMSE (sklearn >= 1.4)

class MLPRegressor(nn.Module):
    def __init__(self, input_dim: int, hidden=(256,128,64), dropout: float = 0.2):
        super().__init__()
        h = list(hidden)
        layers = []
        in_dim = input_dim
        for hdim in h:
            layers += [nn.Linear(in_dim, hdim), nn.ReLU()]
            layers += [nn.Dropout(dropout) if dropout > 0 else nn.Identity()]
            in_dim = hdim
        layers += [nn.Linear(in_dim, 1)]  # salida continua
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x).squeeze(-1)


class MLPRegressorModule:
    def __init__(self,
                 input_dim: int,
                 hidden=(256,128,64),
                 dropout: float = 0.2,
                 lr: float = 1e-3,
                 weight_decay: float = 1e-5,
                 epochs: int = 100,
                 patience: int = 10,
                 seed: int = 42,
                 batch_train: int = 512,
                 batch_eval: int = 1024,
                 use_scheduler: bool = False,
                 scheduler_patience: int = 3,
                 scheduler_factor: float = 0.5):
        # Configuración del entrenamiento
        self.cfg = dict(hidden=hidden, dropout=dropout, lr=lr,
                        weight_decay=weight_decay, epochs=epochs,
                        patience=patience, seed=seed,
                        batch_train=batch_train, batch_eval=batch_eval,
                        use_scheduler=use_scheduler,
                        scheduler_patience=scheduler_patience,
                        scheduler_factor=scheduler_factor)
        self.input_dim = input_dim
        self.model = None

    def _train_one(self, X_tr, y_tr, X_va, y_va):
        # Entrenamiento para un fold con early stopping por RMSE
        set_seed(self.cfg["seed"])
        model = MLPRegressor(self.input_dim,
                             hidden=self.cfg["hidden"],
                             dropout=self.cfg["dropout"])
        criterion = nn.MSELoss()
        optim = torch.optim.Adam(model.parameters(),
                                 lr=self.cfg["lr"],
                                 weight_decay=self.cfg["weight_decay"])

        # Scheduler opcional sobre RMSE de validación
        if self.cfg["use_scheduler"]:
            scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
                optim, mode="min",
                patience=self.cfg["scheduler_patience"],
                factor=self.cfg["scheduler_factor"],
                verbose=False
            )
        else:
            scheduler = None

        # DataLoaders con batch sizes configurables
        train_dl, val_dl = make_loaders(
            X_tr, y_tr, X_va, y_va,
            batch_train=self.cfg["batch_train"],
            batch_eval=self.cfg["batch_eval"]
        )

        best_state, best_rmse, no_imp = None, np.inf, 0
        for _ in range(1, self.cfg["epochs"] + 1):
            # Fase de entrenamiento
            model.train()
            for xb, yb in train_dl:
                optim.zero_grad()
                loss = criterion(model(xb), yb)
                loss.backward()
                optim.step()

            # Fase de validación
            model.eval()
            preds, trues = [], []
            with torch.no_grad():
                for xb, yb in val_dl:
                    preds.append(model(xb).cpu().numpy())
                    trues.append(yb.cpu().numpy())
            yv = np.concatenate(trues)
            pv = np.concatenate(preds)
            rmse = float(root_mean_squared_error(yv, pv))  # sklearn >= 1.4

            # Paso del scheduler si está activo
            if scheduler is not None:
                scheduler.step(rmse)

            # Early stopping por mejor RMSE
            if rmse < best_rmse:
                best_rmse, best_state, no_imp = rmse, deepcopy(model.state_dict()), 0
            else:
                no_imp += 1
                if no_imp >= self.cfg["patience"]:
                    break

        if best_state is not None:
            model.load_state_dict(best_state)
        return model, best_rmse

    def cross_validate(self, X, y, k: int = 5):
        # Validación cruzada K-Fold por RMSE (promedio y desviación)
        splitter = KFold(n_splits=k, shuffle=True, random_state=self.cfg["seed"])
        rmses, models = [], []
        for tr_idx, va_idx in splitter.split(X):
            X_tr, X_va = X[tr_idx], X[va_idx]
            y_tr, y_va = y[tr_idx], y[va_idx]
            model, rmse = self._train_one(X_tr, y_tr, X_va, y_va)
            rmses.append(rmse)
            models.append(model)
        self.model = models[int(np.argmin(rmses))]  # conservar el mejor fold
        return float(np.mean(rmses)), float(np.std(rmses))

    def fit_full_with_es(self, X_train, y_train, valid_size: float = 0.10):
        # Entrenamiento final con validación interna para early stopping
        X_tr, X_va, y_tr, y_va = train_test_split(
            X_train, y_train, test_size=valid_size, random_state=self.cfg["seed"]
        )
        self.model, _ = self._train_one(X_tr, y_tr, X_va, y_va)
        return self

    def predict(self, X):
        # Predicción continua
        self.model.eval()
        with torch.no_grad():
            preds = self.model(torch.from_numpy(np.asarray(X)).float()).cpu().numpy()
        return preds

    def report(self, X_test, y_test, title: str = "MLP Regressor (TEST)"):
        # Reporte con métricas de regresión
        y_pred = self.predict(X_test)
        r2  = r2_score(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        rmse = float(root_mean_squared_error(y_test, y_pred))  # sklearn >= 1.4

        print(f"\n=== {title} ===")
        print(f"R²   = {r2:.3f}")
        print(f"MAE  = {mae:.3f}")
        print(f"RMSE = {rmse:.3f}")

        return y_pred, {"R2": r2, "MAE": mae, "RMSE": rmse}
```

```python
# ====== bloque carga y preparación: dataset café+clima ======
# Lee el archivo, normaliza nombres de columnas, define el target continuo (fanegas),
# detecta columnas numéricas/categóricas, ejecuta el preprocesamiento y prepara dimensiones.

# Ruta del dataset (ajustar si es necesario)
ruta = "dataset_cafe_clima.xlsx"   # usar ruta absoluta si no está en el mismo directorio
df = pd.read_excel(ruta, sheet_name=0)

# Normalizar nombres: minúsculas y sin espacios sobrantes
df.columns = [c.strip().lower() for c in df.columns]

# Target continuo para REGRESIÓN
target_column = "fanegas"
if target_column not in df.columns:
    raise ValueError(f"No se encontró la columna '{target_column}'. Columnas disponibles: {list(df.columns)}")

# Detección automática de columnas numéricas y categóricas (excluyendo el target)
all_cols = df.columns.tolist()
numeric_columns   = df.select_dtypes(include=[np.number]).columns.tolist()
categoric_columns = [c for c in all_cols if c not in numeric_columns]

if target_column in numeric_columns:   numeric_columns.remove(target_column)
if target_column in categoric_columns: categoric_columns.remove(target_column)

# Parámetros de split
test_size = 0.20
random_state = 42

# Preprocesamiento (OneHot + Escalado) y split train/test
X_train_enc, X_test_enc, y_train, y_test = Data_Preprocess(
    df=df,
    numeric_columns=numeric_columns.copy(),
    categoric_columns=categoric_columns.copy(),
    target_column=target_column,
    test_size=test_size,
    random_state=random_state
)

# Exploración rápida (opcional)
explorer = DataExplorer(df, target_column=target_column)
explorer.show_head()
explorer.show_shape()
explorer.show_stats()
explorer.numeric_histograms()

# Asegurar tipos y dimensiones para modelos
X_train_enc = np.asarray(X_train_enc, dtype=np.float32)
X_test_enc  = np.asarray(X_test_enc,  dtype=np.float32)
y_train = np.asarray(y_train, dtype=np.float32)
y_test  = np.asarray(y_test,  dtype=np.float32)

input_dim = X_train_enc.shape[1]
```

```python
# ====== bloque mlp tuning: random search + K-Fold CV por RMSE (con tiempos) ======
# Mide tiempos de: tuning completo (todas las configuraciones) y entrenamiento final del mejor MLP.

import time

def tune_mlp_hparams(
    X, y, input_dim: int,
    k: int = 5, n_trials: int = 25, seed: int = 42, verbose: bool = True,
    search_space: dict | None = None
):
    # Fijar semilla para reproducibilidad en el muestreo Python
    random.seed(seed)

    # Espacio de búsqueda por defecto (ajústalo según el dataset)
    if search_space is None:
        search_space = {
            "hidden": [(256,128,64), (512,256,128), (256,256,128), (128,128,64), (256,128)],
            "dropout": [0.0, 0.1, 0.2, 0.3],
            "lr": [1e-4, 3e-4, 1e-3, 3e-3],
            "weight_decay": [0.0, 1e-5, 1e-4, 1e-3],
            "epochs": [80, 120, 160],
            "patience": [6, 8, 10, 12],
            "batch_train": [128, 256, 512, 1024],
            "batch_eval": [512, 1024, 2048],
            "use_scheduler": [False, True],
            "scheduler_patience": [2, 3, 4],
            "scheduler_factor": [0.5, 0.3]
        }

    keys = list(search_space.keys())
    best = {"rmse_mean": np.inf, "rmse_std": np.inf, "cfg": None, "model": None}

    # Medición del tiempo total de tuning (todas las configuraciones)
    t0_global = time.time()

    for t in range(1, n_trials + 1):
        # Muestreo aleatorio robusto (evita np.random.choice sobre estructuras heterogéneas)
        cfg = {k: random.choice(search_space[k]) for k in keys}

        # Instanciar módulo con la config muestreada
        mlp = MLPRegressorModule(
            input_dim=input_dim,
            hidden=cfg["hidden"],
            dropout=float(cfg["dropout"]),
            lr=float(cfg["lr"]),
            weight_decay=float(cfg["weight_decay"]),
            epochs=int(cfg["epochs"]),
            patience=int(cfg["patience"]),
            seed=seed,
            batch_train=int(cfg["batch_train"]),
            batch_eval=int(cfg["batch_eval"]),
            use_scheduler=bool(cfg["use_scheduler"]),
            scheduler_patience=int(cfg["scheduler_patience"]),
            scheduler_factor=float(cfg["scheduler_factor"])
        )

        # Evaluación por CV (RMSE) con tiempo por configuración (opcional de imprimir)
        t0_cfg = time.time()
        rmse_mean, rmse_std = mlp.cross_validate(X, y, k=k)
        elapsed_cfg = time.time() - t0_cfg

        if verbose:
            print(f"[{t:02d}/{n_trials}] RMSE={rmse_mean:.2f} ± {rmse_std:.2f} | cfg={cfg} | {elapsed_cfg:.1f}s")

        # Actualizar mejor configuración
        if rmse_mean < best["rmse_mean"]:
            best.update({"rmse_mean": rmse_mean, "rmse_std": rmse_std, "cfg": cfg, "model": mlp})

    mlp_time_tuning = time.time() - t0_global
    print(f"\nTiempo total tuning MLP: {mlp_time_tuning:.2f} s")

    return best, mlp_time_tuning

# Ejecutar tuning con el set de entrenamiento preprocesado
best, mlp_time_tuning = tune_mlp_hparams(
    X_train_enc, y_train,
    input_dim=input_dim,
    k=5, n_trials=25, seed=42, verbose=True
)

print("\n>>> Mejor configuración encontrada:")
for k, v in best["cfg"].items():
    print(f" - {k}: {v}")
print(f"CV RMSE = {best['rmse_mean']:.2f} ± {best['rmse_std']:.2f}")

# Reentrenar con early stopping en todo el TRAIN y medir tiempo de entrenamiento final
t0 = time.time()
best_mlp = best["model"].fit_full_with_es(X_train_enc, y_train, valid_size=0.10)
mlp_time_train = time.time() - t0

y_pred_mlp, metrics_mlp = best_mlp.report(X_test_enc, y_test, title="MLP Regressor (TEST) — tuned")
print(f"Tiempo entrenamiento MLP final: {mlp_time_train:.2f} s")
```
